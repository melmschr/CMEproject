```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import scipy.stats as st
import math

# Uncomment to use high DPI for publication-quality figures.
#sns.set(rc={"figure.dpi":300, 'savefig.dpi':300})

# Parse the dataset.
fn = 'data_2021_09_08.csv'
df = pd.read_csv(fn, index_col=0)
```

```python
# Create bar plots to visualize common answers.

da_str = "Virtual away rotations/information sessions before interview season"
response_order = ["Extremely Unhelpful", "Unhelpful", "Neutral", "Helpful", "Extremely Helpful"]

sns.set_theme(style="darkgrid")
ax = sns.countplot(data=df, x=da_str, color='#46166a', saturation=0.75, order=response_order)
ax.set_xticklabels(response_order, rotation=30)
ax.set_ylabel("Number of responses")
ax.set_xlabel(da_str, fontweight="bold")
plt.show()

# This block of questions has a different response structure.
response_order = ["Strongly disagree", "Disagree", "Neutral", "Agree", "Strongly agree"]
sns.set_theme(style="darkgrid")

for col in list(df.columns[86:96]):
  ax = sns.countplot(data=df, x=col, color='#46166a', saturation=0.75, order=response_order)
  ax.set_xticklabels(response_order, rotation=30)
  ax.set_ylabel("Number of responses")
  ax.set_xlabel(col, fontweight="bold")
  plt.show()
```

```python
# Create confidence intervals.

z_star = st.norm.ppf(.975)

# These are blocks of questions which are grouped by response structure. They
# are named by their columns in the excel spreadsheet.
columns_cj_cs = list(df.columns[86:96])
columns_ct_cx = list(df.columns[96:101])
columns_da_dj = list(df.columns[103:113])
columns_dl_du = list(df.columns[114:124])
columns_dy_ee = list(df.columns[127:134])
columns_eg_el = list(df.columns[135:141])

# Map responses in the form to uncased versions with the "strongly" or "very"
# modifiers stripped. We remove the modifiers because we want to consider all
# "agree" and "disagree" responses as the same category for purposes of the
# confidence interval.

# For questions about agreement.
response_map_agree = {
    "Strongly disagree": "disagree",
    "Disagree": "disagree",
    "Neutral": "neutral",
    "Agree": "agree",
    "Strongly agree": "agree"}

# For questions about helpfulness.
response_map_helpful = {
    "Extremely Unhelpful": "unhelpful",
    "Unhelpful": "unhelpful",
    "Neutral": "neutral",
    "Helpful": "helpful",
    "Extremely Helpful": "helpful"}

# We use this map to computed an average response for some questions, where the
# average is computed using these values for each possible response.
response_points_helpful = {
    "Extremely Unhelpful": 0.0,
    "Unhelpful": 1.0,
    "Neutral": 2.0,
    "Helpful": 3.0,
    "Extremely Helpful": 4.0}

# A few columns have slightly different capitalization.
response_map_caps = {
    "Strongly Disagree": "disagree",
    "Disagree": "disagree",
    "Neutral": "neutral",
    "Agree": "agree",
    "Strongly Agree": "agree"}

# Helper which takes returns the mapped response of x, or None if there is no
# mapped response for x in `rmap`.
def map_with_nan(x, rmap):
  if x in rmap:
    return rmap[x]
  else:
    return None

print("--- CJ through CS ---")
for col in columns_cj_cs:
  responses = df[col].apply(lambda x: map_with_nan(x, response_map_agree))
  n = responses.count()
  print(col)
  print("(n =", n, ")")
  # Using this: http://www.stat.yale.edu/Courses/1997-98/101/catinf.htm
  for response in ["disagree", "neutral", "agree"]:
    p_hat = (responses == response).sum() / n
    s_p = math.sqrt((p_hat) * (1- p_hat) / n)
    print("%10s: %05.3f, (%05.3f - %05.3f)" % (response, p_hat*100, (p_hat - s_p * z_star)*100, (p_hat + s_p * z_star)*100))

print("--- CT through CX ---")

points_ct_cx = []

for col in columns_ct_cx:
  responses = df[col].apply(lambda x: map_with_nan(x, response_map_helpful))
  n = responses.count()
  responses_points = df[col].apply(lambda x: map_with_nan(x, response_points_helpful))  # .fillna(0.0)
  #print(responses_points)
  print(col)
  print("(n =", n, ")")
  print(responses_points.mean())
  points_ct_cx.append(responses_points.mean())
  # Using this: http://www.stat.yale.edu/Courses/1997-98/101/catinf.htm
  for response in ["unhelpful", "neutral", "helpful"]:
    p_hat = (responses == response).sum() / n
    s_p = math.sqrt((p_hat) * (1- p_hat) / n)
    print("%10s: %05.3f, (%05.3f - %05.3f)" % (response, p_hat*100, (p_hat - s_p * z_star)*100, (p_hat + s_p * z_star)*100))

sns.barplot(y=columns_ct_cx, x=points_ct_cx, color='#46166a', saturation=0.75)
plt.show()

print("--- DA through DJ ---")

points_da_dj = []

for col in columns_da_dj:
  responses = df[col].apply(lambda x: map_with_nan(x, response_map_helpful))
  n = responses.count()
  responses_points = df[col].apply(lambda x: map_with_nan(x, response_points_helpful))  # .fillna(0.0)
  #print(responses_points)
  print(col)
  print("(n =", n, ")")
  print(responses_points.mean())
  points_da_dj.append(responses_points.mean())
  # Using this: http://www.stat.yale.edu/Courses/1997-98/101/catinf.htm
  for response in ["unhelpful", "neutral", "helpful"]:
    p_hat = (responses == response).sum() / n
    s_p = math.sqrt((p_hat) * (1- p_hat) / n)
    print("%10s: %05.3f, (%05.3f - %05.3f)" % (response, p_hat*100, (p_hat - s_p * z_star)*100, (p_hat + s_p * z_star)*100))

sns.barplot(y=columns_da_dj, x=points_da_dj, color='#46166a', saturation=0.75)
plt.show()

print("--- DL through DU ---")

points_dl_du = []

for col in columns_dl_du:
  responses = df[col].apply(lambda x: map_with_nan(x, response_map_helpful))
  n = responses.count()
  responses_points = df[col].apply(lambda x: map_with_nan(x, response_points_helpful))
  print(col)
  print("(n =", n, ")")
  print(responses_points.mean())
  points_dl_du.append(responses_points.mean())
  # Using this: http://www.stat.yale.edu/Courses/1997-98/101/catinf.htm
  for response in ["unhelpful", "neutral", "helpful"]:
    p_hat = (responses == response).sum() / n
    s_p = math.sqrt((p_hat) * (1- p_hat) / n)
    print("%10s: %05.3f, (%05.3f - %05.3f)" % (response, p_hat*100, (p_hat - s_p * z_star)*100, (p_hat + s_p * z_star)*100))


sns.barplot(y=columns_dl_du, x=points_dl_du, color='#46166a', saturation=0.75)
plt.show()

print("--- DY through EE ---")

for col in columns_dy_ee:
  responses = df[col].apply(lambda x: map_with_nan(x, response_map_caps))
  n = responses.count()
  print(col)
  print("(n =", n, ")")
  # Using this: http://www.stat.yale.edu/Courses/1997-98/101/catinf.htm
  for response in ["disagree", "neutral", "agree"]:
    p_hat = (responses == response).sum() / n
    s_p = math.sqrt((p_hat) * (1- p_hat) / n)
    print("%10s: %05.3f, (%05.3f - %05.3f)" % (response, p_hat*100, (p_hat - s_p * z_star)*100, (p_hat + s_p * z_star)*100))

print("--- EG through EL ---")

for col in columns_eg_el:
  responses = df[col].apply(lambda x: map_with_nan(x, response_map_caps))
  n = responses.count()
  print(col)
  print("(n =", n, ")")
  # Using this: http://www.stat.yale.edu/Courses/1997-98/101/catinf.htm
  for response in ["disagree", "neutral", "agree"]:
    p_hat = (responses == response).sum() / n
    s_p = math.sqrt((p_hat) * (1- p_hat) / n)
    print("%10s: %05.3f, (%05.3f - %05.3f)" % (response, p_hat*100, (p_hat - s_p * z_star)*100, (p_hat + s_p * z_star)*100))
  
  
```
